```python
!pip3 install yfinance
import yfinance as yf
import pandas as pd

# List of stock symbols
stock_symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'FB', 'TSLA', 'NFLX', 'NVDA', 'INTC', 'AMD', 'CSCO', 'IBM', 'ORCL', 'HPQ', 'QCOM']

# Define the date range
start_date = '2018-10-23'
end_date = '2023-10-21'

# Create an empty DataFrame to store the data
all_data = pd.DataFrame()

# Download historical data for each stock
for stock_symbol in stock_symbols:
    data = yf.download(stock_symbol, start=start_date, end=end_date)
    if not data.empty:
        data['Symbol'] = stock_symbol  # Add a column to identify the stock
        all_data = pd.concat([all_data, data], axis=0)

# Display the first few rows of the combined data
print(all_data.head())

# Save the combined data to a CSV file
csv_filename = 'multiple_stocks_historical_data.csv'
all_data.to_csv(csv_filename)

print(f'Data saved to {csv_filename}')
```
